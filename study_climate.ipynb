# Study of Temperature of Nepal year 2020 and 2050

```python
# Load the raster data for temperature in Nepal for 2020 and 2050
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the .tif files
file1_path = "C:/Sujal_Limbu_Assignment/Python/GIS/gis-data-science-assignment-sujal-limbu/Data/nepal_temperature_2020.tif"
file2_path = "C:/Sujal_Limbu_Assignment/Python/GIS/gis-data-science-assignment-sujal-limbu/Data/nepal_temperature_2050 (1).tif"

with rasterio.open(file1_path) as src1, rasterio.open(file2_path) as src2:
    data_2020 = src1.read(1)
    data_2050 = src2.read(1)
    meta1 = src1.meta

# Step 2: Compute the difference
difference = data_2050 - data_2020

# Step 3: Statistical Summaries
def summarize_temperature(data, year):
    print(f"Statistics for {year}:")
    print(f"  Mean temperature: {np.nanmean(data):.2f} °C")
    print(f"  Median temperature: {np.nanmedian(data):.2f} °C")
    print(f"  Minimum temperature: {np.nanmin(data):.2f} °C")
    print(f"  Maximum temperature: {np.nanmax(data):.2f} °C")
    print(f"  Standard deviation: {np.nanstd(data):.2f} °C")
    print()

summarize_temperature(data_2020, "2020")
summarize_temperature(data_2050, "2050")

# Step 4: Visualizations
# Visualization 1: Temperature Comparison (2020 vs 2050)
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(data_2020, cmap="plasma")
plt.colorbar(label="Temperature 2020 (°C)")
plt.title("Temperature 2020")

plt.subplot(1, 3, 2)
plt.imshow(data_2050, cmap="plasma")
plt.colorbar(label="Temperature 2050 (°C)")
plt.title("Temperature 2050")

plt.subplot(1, 3, 3)
plt.imshow(difference, cmap="coolwarm", vmin=-5, vmax=5)
plt.colorbar(label="Temperature Change (°C)")
plt.title("Temperature Change (2050 - 2020)")

plt.tight_layout()
plt.show()

# Visualization 2: Histogram of Temperature Distribution
plt.figure(figsize=(10, 6))
plt.hist(data_2020.flatten(), bins=30, alpha=0.5, label="2020", color="blue")
plt.hist(data_2050.flatten(), bins=30, alpha=0.5, label="2050", color="red")
plt.title("Temperature Distribution (2020 vs 2050)")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()

# Visualization 3: Boxplot of Temperature
plt.figure(figsize=(8, 6))
plt.boxplot([data_2020.flatten(), data_2050.flatten()], labels=["2020", "2050"])
plt.title("Temperature Comparison (2020 vs 2050)")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()

# Step 5: Analyze Trends
print("Mean temperature change:", np.nanmean(difference))
print("Max temperature increase:", np.nanmax(difference))
print("Max temperature decrease:", np.nanmin(difference))

# Step 6: Save the difference
output_path = "temperature_difference_2020_2050.tif"
with rasterio.open(output_path, "w", **meta1) as dst:
    dst.write(difference, 1)
```

Key Insight for Temperature of Nepal


- Widespread Warming:

   1) The mean temperature increase of 2.10°C aligns with global climate change trends, indicating that Nepal is experiencing significant warming.

   2) This warming is consistent with the impacts of greenhouse gas emissions and global climate change.

- Regional Variability:

   1) The maximum temperature increase of 5.46°C highlights that some regions are warming at a much faster rate than others.

   2) Areas with the highest increases could be lowland regions (e.g., Terai) or urban areas experiencing the urban heat island effect.

-  Localized Cooling:

   1) The maximum temperature decrease of -1.12°C suggests that some areas may have experienced slight cooling.

   2) This could be due to factors such as changes in land use, cloud cover, or localized climatic phenomena.
